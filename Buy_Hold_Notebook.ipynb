```
# Creating function that finds max or min value in specified range, and the row
def get_max_min(start, end, direction):

  date_range = df[(df.index >= start) & (df.index <= end)]
  if direction == 'min':
    max_or_min_value = date_range['Real Price'].min()
  elif direction == 'max':
    max_or_min_value = date_range['Real Price'].max()
  else:
    print("Must enter 'min' or 'max'")
  
  return df[df['Real Price'] == max_or_min_value]

get_max_min(dt.datetime(1871, 8, 1), dt.datetime(1906, 9, 1), 'min')
```

```
# test date
#date = dt.datetime(1900, 1, 1)


# Found add_years function on stackoverflow: https://stackoverflow.com/questions/15741618/add-one-year-in-current-date-python
def add_years(d):
  
  """Return a date that's `years` years after the date (or datetime)
  object `d`. Return the same calendar date (month and day) in the
  destination year, if it exists, otherwise use the following day
  (thus changing February 29 to March 1).

  """
  years = 40
  try:
      return d.replace(year = d.year + years)
  except ValueError:
      return d + (date(d.year + years, 1, 1) - date(d.year, 1, 1))
      
#add_years(date)
```
